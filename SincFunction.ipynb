```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
@np.vectorize
def sinc(x):
    if x == 0:
        return 1.
    else:
        return np.sin(x) / x
```

```python
sinc(np.array([-0.1, 0, 0.1, 1., 1.5]))
```

```python
x = np.linspace(-20, 20, 10001)
```

```python
y = sinc(x)
```

```python
plt.scatter(x, y, s=0.1)
```

```python
targety = y + np.random.normal(0, 0.025, size=y.shape[0])
```

```python
plt.scatter(x, targety, s=0.1)
```

```python
np.save('x', x)
np.save('y', targety)
```
