```python
import numpy as np
import tensorflow as tf
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.load('x.npy')
y = np.load('y.npy')
```

```python
model1 = tf.keras.Sequential()
model1.add(tf.keras.layers.Dense(16, activation='relu', input_shape=(1,)))
model1.add(tf.keras.layers.Dense(1))
model1.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model1.summary()
```

```python
model1.fit(x, y, epochs=1000)
```

```python
model1.save('tfsincmodel')
```

```python
testx = np.linspace(-10, 10, 157)
```

```python
testy = model1.predict(testx)
```

```python
plt.scatter(testx, testy, s=0.1)
```
